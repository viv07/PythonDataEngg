# join emp and dept data in query

```python
edquery= 'select e.empno, e.ename, e.sal, d.deptno, d.dname from emp_df e inner join dept_df d on e.deptno=d.deptno'
merge_df=dftosql(edquery)
print(type(merge_df))
merge_df.head(10)
```

# groupby and aggregation

```python
#calc deptwise avg salary
agg_query='select deptno, round(avg(sal),2) as avg from emp_df group by deptno'
agg_df=dftosql(agg_query)
print(type(agg_df))
agg_df.head()
```
